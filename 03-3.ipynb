```python
import pandas as pd
```

```python
df = pd.read_csv('data/fish.csv')
df
```

```python
perch_df = df.loc[df['Species']=='Perch']
perch_df.head()
```

```python
perch_full = perch_df[['Length2','Height','Width']]
perch_weight = perch_df[['Weight']]
```

```python
perch_full.head()
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(perch_full, perch_weight, random_state = 42)
```

```python
from sklearn.preprocessing import PolynomialFeatures
```

```python
poly = PolynomialFeatures()
poly.fit([[3,5]])
poly.transform([[3,5]])
```

```python
poly = PolynomialFeatures(include_bias=False)
poly.fit(X_train)
train_pol = poly.transform(X_train)
```

```python
train_pol
```

- 릿지 : 계수를 제곱한 값을 기준으로 규제 
  - L2 norm : 모든 계수를 작게 만들지만 0은 되지 않음
- 라쏘:  계수의 절댓값을 기준으로 규제 
  - L1 norm : 불필요한 특성의 가중치를 0으로 만들음
