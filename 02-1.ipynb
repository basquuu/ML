```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('data/fish.csv')
df
```

```python
df_Bream = df[df['Species'] =='Bream']
df_Smelt = df[df['Species'] =='Smelt']
bream_length = df_Bream['Length2'].tolist()
bream_weight = df_Bream['Weight'].tolist()
smelt_length = df['Length2'].tolist()
smelt_weight = df['Weight'].tolist()
```

```python
length = bream_length + smelt_length
weight = bream_weight + smelt_weight
fish_data = [[l,w] for l, w in zip(length, weight)]
target_data = [1] * 35 + [0] * 14
```

```python
fish_data
```

### train_test split

```python
from sklearn.model_selection import train_test_split
X = fish_data
y = target_data

X_train, X_test, y_train, y_test = train_test_split(X,y, random_state=42,shuffle=True)
```
